# Collections

````python
# WRITE A PROGRAM TO find to count of each char ( frequency count of your data)
# input : lavanya
# output : {l:1, a:3, v:1, n:1, y:1}
````

````python
def fCount(n):
    res={}
    for i in n:
        if i in res:
            res[i]+=1
        else:
            res[i]=1
    return res   
fCount("Lavanya")
````
