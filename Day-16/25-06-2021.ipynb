#### Outliers
- Outliers are different from other data points.
- If dataset contains outliers, it will effect the training process of machine learning models and will face loss of accuracy

```python
df = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv')
```

```python
df.head()
```

### 1. Standard Scaler
- By applying Standard scaler, the dataset will be normally distributed
- It scales the data point between mean to unit variance
    - **Formula**:
        - Z = (Xi - mean(X) / std(X)

```python
plt.hist(df)
plt.show()
```

```python
df['TV'].mean()
```

```python
df['TV'].std()
```

```python
(230.1 - 147.0425) / 85.8542363149081
```

#### Apply standard scaler model

```python
st2 = st.fit_transform(df)
st2
```

### Robust Scaler
- If the dataset contains large amount of outliers, then use Robust scaler
- It scales the data points between Inter quartile range (25% to 75%)
    - Lower quartile(q1) = 25%
    - Mediun (q2) = 50%
    - Upper quartile (q3) = 75%
- **Formula:**
    - (Xi - median(X)) / (Q3(X) - Q1(X))

```python
df.head()
```

```python
df['TV'].median()
```

```python
df['TV'].quantile(0.25)
```

```python
df['TV'].quantile(0.75)
```

```python
(230.1 - 149.75) / (218.825 - 74.375)
```

#### Apply Robust scaler model

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
```

```python
rs2 = rs.fit_transform(df)
rs2
```

```python
rs2 = pd.DataFrame(rs2, columns = df.columns)
rs2
```
