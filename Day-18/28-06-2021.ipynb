### 1. Palette
- It is an interface to generate different shades colors
-  Types
    - Qualitative
    - Sequential
    - Diverging
    
    
- sns.color_palette()

#### Qualitative: Palette contains multiple colors
    - Deep
    - Pastel
    - Dark
    - muted
    - Bright
    - Colorblind

```python
help(sns.color_palette)
```

```python
sns.palplot(sns.color_palette())
```

```python
sns.palplot(sns.color_palette('dark'))
```

```python
sns.palplot(sns.color_palette('pastel'))
```

#### Sequential: Palette contains only one color

```python
sns.palplot(sns.color_palette('Greens'))
```

```python
sns.palplot(sns.color_palette('Blues'))
```

```python
sns.palplot(sns.color_palette('Reds'))
```

#### Diverging: Palette contains 2 types of colors

```python
sns.palplot(sns.color_palette('RdBu'))
```

```python
sns.palplot(sns.color_palette('PuOr'))
```

```python
sns.palplot(sns.color_palette('Purp'))
## Error and Availabe palette colors
```

```python
sns.palplot(sns.color_palette('RdBu',n_colors = 10))
```

```python
sns.palplot(sns.color_palette('Reds',n_colors = 100))
```

### 2. Categorical Plots
- By default, it return scatterplot
    - sns.catplot()

```python
help(sns.catplot)
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = iris)
```

- **Boxplot**

```python
sns.catplot(data = iris, kind = 'box', orient = 'h')
```

```python
sns.catplot(x = 'species', y = 'petal_length', data = iris, kind = 'box', palette = 'Purple')
```

```python
sns.catplot(x = 'species', y = 'petal_length', data = iris, kind = 'box', palette = 'YlOrRd')
```

- **Violin Plot**
    - Similar to boxplot

```python
sns.catplot(x = 'species', y = 'petal_length', data = iris, kind = 'violin', palette = 'gnuplot')
```

- **Strip Plot**
    - Used when one of the column is categorical. It represented as the date in sorted order

```python
sns.catplot(x = 'species', y = 'sepal_width', data = iris, kind = 'strip')
```

- **Swarm Plot**
    - Generally swarm means insect. In seaborn swarm plot is also known as 'bees plot'.
    - Used to avoid of overlapping

```python
sns.catplot(x = 'species', y = 'petal_width', data = iris, kind = 'swarm')
```
