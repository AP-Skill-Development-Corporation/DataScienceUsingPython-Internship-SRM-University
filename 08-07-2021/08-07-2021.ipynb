### Random Forest Model
- Random forest algorithm is one of the Ensemble Technique.
- Ensemble method will split data into multiple parts and then combine them to produce improved results.
- Collection of Decision tree models is called a Random Forest Model. (Random forest is similar to Decision tree.)
- Random Forest plays a key role for huge dataset. (lakhs of rows and 100s of columns)
- Applicable for both classification and regression model

### Working of Random Forest model
- Step1: Selection of random samples in given dataset
- Step2: It will apply Decision Tree model for every sample.
- Step3: It will get the prediction result for every decision tree. And voyeing will be performed.
- Step4: Majority voted prediction result will be taken as final prediction result.
- Eg:
    - Training Sample1 --> Female
    - Training Sample2 --> Male
    - Training SampleN --> Male
    - 'Majority voted is Male'

```python
X = df.drop('Outcome', axis = 1) ## Input variable
Y = df['Outcome']    ## Target variable
```

```python
X.head()
```

```python
Y[:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.30, random_state = 45)
```

```python
X_train
```

```python
X_train.shape , X_test.shape
```

```python
Y_train.shape, Y_test.shape
```

```python
### Apply Random forest model
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc
```

```python
rfc.fit(X_train, Y_train)
```

```python
Y_prediction = rfc.predict(X_test)
Y_prediction
```

```python
X_test.head(3)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_prediction, Y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_prediction, Y_test)
```
