### Distribution Plot
- It plots the univariate distribution of data.

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.distplot(iris['sepal_length'])
## KDE --> Kernel Density Estimation
## Estimates probability density of curve which have 1 or more continous variables
```

```python
help(sns.distplot)
```

```python
sns.distplot(iris['sepal_length'], kde = False)
```

- **Seaborn have Built-In-Styles**
    - Darkgrid
    - Whitegrid
    - Drak
    - White
    - Ticks
- sns.set_style()

```python
sns.set_style('darkgrid')
sns.distplot(iris['sepal_length'], vertical = True)
```

### Joint Plot
- Combination of 2 plots. (By default scatter and histogram plots)

```python
iris.head(2)
```

```python
sns.jointplot(x = 'petal_length', y = 'sepal_width', data = iris, color = 'g')
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x = 'petal_length', y = 'sepal_width', kind = 'hex', data = iris, color = 'orange')
```

```python
sns.jointplot(x = 'petal_length', y = 'sepal_width', kind = 'kde', data = iris, color = 'm')
```

### Pair plot
- Used to visualize pairwise relationship between columns

```python
sns.pairplot(iris)
## Diagnally --> Column is pairing with itself
```

```python
sns.pairplot(iris, hue = 'species')
```

```python
help(sns.pairplot)
```

```python
sns.pairplot(iris, hue = 'species', vars = ['sepal_length','petal_length'])
```

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
```

```python
sns.pairplot(iris, hue="species", vars = ['sepal_length','petal_length'], 
             markers=["+", "h", "D"], palette = 'gnuplot_r')
```

### Heatmap
- Heatmap visualize as matrix-like data and checks how strongly each column is depended on other columns

```python
tips.head()
```

```python
sns.heatmap(tips)
```

```python
tips.isnull().sum()
```

```python
tips.iloc[:,[0,1,-1]].head()
```

```python
sns.heatmap(tips.iloc[:,[0,1,-1]])
```

```python
sns.heatmap(tips.iloc[:,[0,1,-1]].head(), annot = True)
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(), annot = True)
```

```python
help(sns.heatmap)
```

```python
sns.heatmap(tips.corr(), annot = True, cmap = 'Oranges_r', linewidths = 2, linecolor='pink')
```

### lmplot
- Combination of Line and scatter

```python
tips.head()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
help(sns.lmplot)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, scatter=False)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue ='smoker', palette = 'Set1')
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",data=tips, markers=["D", "x"], palette = 'gnuplot_r')
```

```python
sns.lmplot(x="total_bill", y="tip", hue = 'time',data=tips, col = 'smoker', palette = 'gnuplot_r')
```
