```python
x = np.arange(1,100,5)
y = x**3
plt.plot(x,y, c = "g", label = "(x, x**3)")
plt.plot(x*2,y, "b", label = ("x*2, x**3"))
plt.xlabel("X-axis")
plt.ylabel("y-Axis")
plt.grid()
plt.legend(loc = "upper left")
plt.show()
```

```python
x = np.arange(1,100,5)
y = x**3
plt.plot(x,y, c = "g", label = "(x, x**3)", linestyle = "-.")
# supported values for line style are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.plot(x*2,y, "b", label = ("x*2, x**3"), linestyle = ":")
plt.xlabel("X-axis")
plt.ylabel("y-Axis")
plt.grid()
plt.legend(loc = "upper left")
plt.show()
```

```python
x = np.arange(1,100,5)
y = x**3
plt.plot(x,y, c = "g", label = "(x, x**3)", linestyle = "-.", marker = "*")
# supported values for line style are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.plot(x*2,y+3, "b", label = ("x*2, x**3"), linestyle = ":", marker = ">")
plt.xlabel("X-axis")
plt.ylabel("y-Axis")
plt.grid()
plt.legend(loc = "upper left")
plt.show()
```

# Subplots


```python

fig, area = plt.subplots(2,3)
plt.show()
```

```python
fig , area = plt.subplots(2,2)
x = np.arange(1,10)
area[0][0].plot(x,x)
area[0][0].set_title("Line")
area[0][1].plot(x,x**2)
area[0][1].set_title("Square")
area[1][0].plot(x,x**3)
area[1][0].set_title("Cube")
area[1][1].plot(x,np.log(x))
area[1][1].set_title("Log")
plt.show()
```

```python
g1 = plt.subplot2grid((3,3),(0,0), colspan = 2)
g2 = plt.subplot2grid((3,3),(0,2), rowspan = 2)
g3 = plt.subplot2grid((3,3),(1,0))
```

```python
plt.figure(figsize = (7,7))
g1 = plt.subplot2grid((3,3),(0,0),colspan = 2)
g2 = plt.subplot2grid((3,3),(0,2), rowspan = 2)
g3 = plt.subplot2grid((3,3),(1,0), colspan = 2, rowspan = 1)
x = np.arange(1,20)
g1.plot(x,np.power(x,2), c = "c")
g2.plot(x,np.exp(x), c ="black")
g3.plot(x,np.sin(x), c = "y")
plt.show()
```

# Pie Chart

```python
x = ["C","Java","Python","R","c#"]
y = [90,45,100,87,67]
plt.pie(y, labels = x, colors = ["r","g","b","m","c"], radius = 1.5, autopct = "% 1.2f %%", shadow = True)
plt.show()
```

```python
x = ["C","Java","Python","R","c#"]
y = [90,45,100,87,67]
plt.pie(y, labels = x, colors = ["r","g","b","m","c"], radius = 1, explode =[0,0,1,0,0] )
plt.show()
```

```python
x = ["C","Java","Python","R","c#"]
y = [90,45,100,87,67]
plt.pie(y, labels = x, colors = ["r","g","b","m","c"], radius = 1, explode =[0,1,1,1,0], autopct= "% 1.1f %%" )
plt.show()
```
