### Support Vector Machine Classification Model
- This model is used to find the hyperplane between 2 classes in dataset.
- Hyperplane is a line which linearly seperates the classes.

### Maximum Width
- w * x - b = +1
- w * x - b = 0
- w * x - b = -1



```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/lung_cancer_examples.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
## Seperate Input columns and Target columns
X = df[['Age','Smokes', 'AreaQ','Alkhol']]
Y = df['Result']
```

```python
X.head()
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
svm
```

```python
svm = SVC(kernel='linear')
svm
```

```python
svm.fit(X,Y)
```

```python
Y_prediction = svm.predict(X)
Y_prediction
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y, Y_prediction)
```

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Student_PassOrFail.csv')
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2.isnull().sum()
```

```python
X2 = df2[['Self_Study_Daily','Tution_Monthly']]
Y2 = df2['Pass_Or_Fail']
```

```python
from sklearn.svm import SVC
```

```python
svm2 = SVC(kernel='linear')
svm2
```

```python
svm2.fit(X2,Y2)
```

```python
Y_prediction2 = svm2.predict(X2)
Y_prediction2
```

```python
accuracy_score(Y2, Y_prediction2)
```

```python
svm3 = SVC(kernel='poly')
svm3
```

```python
svm3.fit(X2,Y2)
```

```python
Y_prediction3 = svm3.predict(X2)
```

```python
accuracy_score(Y2, Y_prediction3)
```
